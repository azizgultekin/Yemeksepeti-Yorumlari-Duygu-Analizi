```python
from bs4 import BeautifulSoup
import requests

r = requests.get("https://www.yemeksepeti.com/burger-king-istanbul?source=home#ors:true")
soup = BeautifulSoup(r.content,"lxml")
st1 = soup.find("div",attrs={"class":"col-md-12 ys-rightColumn"})
st2 = st1.find("div",attrs={"class":"ys-reslist"})
st3 = st2.find_all("div",attrs={"class":"ys-item"})

for icerik in st3:
    st4=icerik.find("div",attrs={"class":"restaurant-display-name"})
    link_sonu = st4.a.get("href")
    link_basi = "https://www.yemeksepeti.com"
    link = link_basi+link_sonu
    a=1
    x= str(link)+"?section=comments&page="
    r2 = requests.get(x+str(a))
    soup1 = BeautifulSoup(r2.content,"lxml")
    yorum_sayisi = soup1.find("a",attrs={"href" : "?section=comments"}).text
    dizi=yorum_sayisi.split(" ")
    ab=dizi[1].replace("(","")
    ca=ab.replace(")","")
    da=int(ca)/30

    dosya = open("KFC_Yorumlari.txt","a",encoding="utf-8")

    for dongu in range(0,int(da)+1):

        r3 = requests.get(x+str(a))
        soup2 = BeautifulSoup(r3.content,"lxml")
        st5 = soup2.find("div",attrs={"class":"comments allCommentsArea"})
        st6 = st5.find_all("div",attrs={"class":"comments-body"})
        for icerik in st6:
                st7 = icerik.find("div",attrs={"class":"user"})

                if ( st7.find("div",attrs={"class":"comment ys-comment row"}) in st7):
                    print()
                    #st4 = st3.find("div",attrs={"class":"comment ys-comment row"})
                    #dosya.write(st4.find("p").text+"\n")

                else:
                    st8 = st7.find("div",attrs={"class":"comment row"})
                    dosya.write(st8.find("p").text+"\n")
        a += 1      

    dosya.close()
```
